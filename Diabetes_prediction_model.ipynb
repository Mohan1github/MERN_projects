```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
```

```python
#Data processing and data cleaning and loading
data_set = pd.read_csv("/content/sample_data/diabetes.csv")
print(data_set.shape)
print(data_set.head())
print(data_set.tail())
```

```python
null = data_set.isnull().sum()
print(null)
```

```python
data_set.describe()
```

```python
data_set['Outcome'].value_counts()
```

```python
data_set.groupby('Outcome').mean()
```

```python
x = data_set.drop(columns = 'Outcome',axis = 1)
y = data_set['Outcome']
print(x)
print(y)
```

```python
#transform the data into standard data
scaler = StandardScaler()
standard_data = scaler.fit_transform(x)
print(standard_data)
```

Data spliting

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify =y, random_state = 2)
print(x_train)
print(x_test)
```

training the model

```python
classifier = svm.SVC(kernel = 'linear')
classifier.fit(x_train,y_train)
```

```python
x_train_prediction = classifier.predict(x_train)
accuracy_x = accuracy_score(x_train_prediction,y_train)
print(accuracy_x)
```

```python
x_test_prediction = classifier.predict(x_test)
accuracy_x_test = accuracy_score(x_test_prediction,y_test)
print(accuracy_x_test)
```

predictive system


```python
input_data =(1,89,66,23,94,28.1,0.167,21)
input_np = np.asarray(input_data)
input_reshaped = input_np.reshape(1,-1)
standard_data_predictive = scaler.fit_transform(input_reshaped)
print(standard_data_predictive)
prediction = classifier.predict(standard_data_predictive)
print(prediction)
if(prediction[0] == 0):
  print("This person don't have diabetes")
else:
  print("This person have diabetes")
```
